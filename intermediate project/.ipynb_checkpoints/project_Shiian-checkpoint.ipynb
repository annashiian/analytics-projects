```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df.head()
```

```python
df['purchaseYMW']=pd.to_datetime(df.order_purchase_timestamp).dt.strftime('%Y-%m')
```

```python
groups=df.groupby(['customer_unique_id','purchaseYMW'],as_index=False)\
    .agg({'order_id':'nunique'})\
    .sort_values('purchaseYMW')
```

```python
groups.set_index('customer_unique_id',inplace=True)
```

```python
groups['cohort']=groups.groupby(level=0)['purchaseYMW'].min()
```

```python
groups.reset_index(inplace=True)
```

```python
order_sum=groups.groupby(['cohort','purchaseYMW'])\
    .agg({'order_id':'sum'})
```

```python
#объявляем функцию, которая приводит дату(строка) к datetime

def parse_date(x):
    
    return pd.to_datetime(x, format="%Y-%m")

#объявляем функцию,которая вычисляет разницу между датами
def month_diff(a, b):
    
    a = parse_date(a)
    b = parse_date(b)
    return 12 * (a.year - b.year) + (a.month - b.month)

#объявляем функцию, которая для каждой когорты считает разницу от начала когорты(первого месяца покупки) до (каждого)месяца покупки
def cohort_period(df):
    
    df['CohortPeriod']=month_diff(df.index.get_level_values('purchaseYMW'), df.index.get_level_values('cohort'))
    df['order_id'] = df['order_id']/df['order_id'][0]
    return df
order_sum=order_sum.groupby(level=0).apply(cohort_period)
```

```python
#создаем таблицу
retentation_rate=order_sum.reset_index().pivot(index='cohort',columns='CohortPeriod',values='order_id')
```

```python
#преобразуем в "тепловую карту" (стиль)
ur_style = (retentation_rate
            .style
            .set_caption('User retention by cohort')  # добавляем подпись
            .background_gradient(cmap='viridis')  # раскрашиваем ячейки по столбцам
            .highlight_null('white')  # делаем белый фон для значений NaN
            .format("{:.2%}", na_rep=""))  # числа форматируем как проценты, NaN заменяем на пустоту
ur_style
```

```python
retention_rate_january_december=retention_rate_jan_dec[retention_rate_jan_dec.index<='2017-12']
```

```python
#убедимся, что  that customer unique id and order id ,order_purchase_timestamp не содержат na
df.isna().sum()
```

```python
# уберем все na в order_approved_at чтобы оставить только оплаченные заказы
rfm_data=df.dropna(subset=['order_approved_at'])
```

```python
#приведем столбец order_approved_at к нужному типу
rfm_data['order_approved_at'] = pd.to_datetime(rfm_data['order_approved_at'])
```

```python
#order_approved_at считаем моментом покупки. определим временные рамки данных
print('Orders from {} to {}'.format(rfm_data['order_approved_at'].min(), rfm_data['order_approved_at'].max()))
```

```python
# В одном заказе может быть несколько товаров.
# Нас интересует количество заказов которые сделал каждый покупатель, а не количество покупок которые они делают
# Сгруппируем все данные по уникальным заказам. Найдем суммарную стоимость для каждого
orders=rfm_data.groupby(['customer_unique_id', 'order_id', 'order_approved_at'], as_index=False).agg({'price': 'sum'})
```

```python
# Для удобства анализа мы предположим, что мы собрали данные на сегодняшний день. Таким образом день последней оплаты обозначим как сегодняшний
NOW = orders.order_approved_at.max()
NOW
```

```python
# посчитаем время от последней покупки до текущей даты
orders['days_since_last_order'] = orders['order_approved_at'].apply(lambda x: (NOW - x).days)
orders.head()
```

```python
# рассчитываем для каждого пользователя: R - кол-во дней с последнего заказа и F - количество заказов за все время и 
# М - потраченную сумму за все заказы
rfm=orders.groupby(['customer_unique_id'], as_index=False)\
    .agg({\
        'days_since_last_order':'min',\
        'order_id': 'nunique',\
        'price':'sum'\
    })\
    .rename(columns={'days_since_last_order':'R', 'order_id':'F', 'price':'M'})
rfm.head()
```

```python
# большинство людей сделали всего один заказ
rfm.F.value_counts()
```

```python
# для расчета рейтингов каждого показателя, разделим все данные на группы по каждому из показателей
quantilies = rfm[['R', 'F', 'M']].quantile([.33, .66]).to_dict()
```

```python
# напишем функции для определения рейтингов. чем меньше R тем лучше и наоборот, чем больше F и M тем лучше
def r_score(x):
    if x <= quantilies['R'][.33]:
        return 3
    elif x <= quantilies['R'][.66]:
        return 2
    else:
        return 1

def f_score(x):
    if x <= quantilies['F'][.33]:
        return 1
    elif x <= quantilies['F'][.66]:
        return 2
    else:
        return 3
    
def m_score(x):
    if x <= quantilies['M'][.33]:
        return 1
    elif x <= quantilies['M'][.66]:
        return 2
    else:
        return 3
```

```python
rfm['R_sc']=rfm['R'].apply(r_score)
rfm['F_sc']=rfm['F'].apply(f_score)
rfm['M_sc']=rfm['M'].apply(m_score)
rfm.head()
```

```python
rfm['RFM Score'] = rfm['R_sc'].map(str) + rfm['F_sc'].map(str) + rfm['M_sc'].map(str)
rfm.head()
```
